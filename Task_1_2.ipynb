    : 
\begin{equation}
\lim_{n  \rightarrow +\infty} \frac{n}{\sqrt[n]{n!}} = \lim_{n  \rightarrow +\infty} \sqrt[n]{\frac{n^n}{n!}} = \lim_{n  \rightarrow +\infty} \sqrt[n]{a_n},
\end{equation}
 $a_n = \frac{n^n}{n!}$.  $a_n > 0$    : 
\begin{equation}
\lim_{n  \rightarrow +\infty} \sqrt[n]{a_n} = \lim_{n  \rightarrow +\infty} \frac{a_{n+1}}{a_n}
\end{equation}
  : 
\begin{equation}
\lim_{n  \rightarrow +\infty} \frac{n}{\sqrt[n]{n!}} = \lim_{n  \rightarrow +\infty} \sqrt[n]{\frac{n^n}{n!}} = \lim_{n  \rightarrow +\infty} \frac{(n+1)^{n+1}}{(n+1)!} \frac{n!}{n^n} = \lim_{n  \rightarrow +\infty} \frac{(n+1)^n}{n^n} = e,
\end{equation}

```python
%%time

import matplotlib
import math
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update({'font.size': 12})

import math

def series_n(n):
    return (1 + 1/n)**n

epsilon = 1e-7
n = 1
while abs(series_n(n + 1) - series_n(n)) > epsilon:
    n = n + 1
    
print(' n:', n, '\n a_n:', series_n(n))
```

```python
%%time

import matplotlib
import math
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update({'font.size': 12})


def series_n(n):
    return math.pow((1 + 1/n), n)

epsilon = 1e-7
n = 1
root_x = [series_n(1)] # 
root_y = [0]           # 
while abs(series_n(n + 1) - series_n(n)) > epsilon:
    n = n * 2
    root_x.append(series_n(n)) # 
    root_y.append(0)           # 
    
while abs(series_n(n + 1) - series_n(n)) < epsilon:
    n = n - 1

print(' n:', n + 1, '\n a_n:', series_n(n))
```
