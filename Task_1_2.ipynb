## Тема 3 “Последовательность”

### Задание 3.1

Исследовать на монотонность, огранниченность и найти 5 член по счету. 

**Заданине 3.1а**
\begin{equation}
\begin{gathered}
\quad\left\{a_n\right\}_{n=1}^{\infty} = 2^n-n
\end{gathered}
\end{equation}
Рассморим $n+1$ элемент последовательнсоти:
\begin{equation}
\begin{gathered}
a_{n+1} = 2^{n+1}-n-1 = 2\cdot 2^{n}-2n-1+n = 2a_n + n -1
\end{gathered}
\end{equation}
Из полученного выражения видно, что последовательность монотонно возрастает (так как величина $a_{n+1} > a_n$).

Данная последовательность ограничена снизу (при $n=1$) числом 1, сверху данная последовательность неограничена. Отсюда следует, что в целом $a_n$ неограничена.

\begin{equation}
a_{5} = 2^5-5 = 27
\end{equation}

**Заданине 3.1b**
\begin{equation}
\begin{gathered}
\quad\left\{b_n\right\}_{n=2}^{\infty} = \frac{1}{1-n}
\end{gathered}
\end{equation}
Рассморим разность между $b_{n+1}$ и $b_n$:
\begin{equation}
b_{n+1} - b_n = \frac{1}{1 - (n-1)} - \frac{1}{1-n} = -\frac{1}{n} - \frac{1}{1-n} = \frac{1}{n(n-1)} > 0
\end{equation}
Из полученного выражения видно, что последовательность монотонно возрастает.

Данная последовательность ограничена снизу (при $n=2$) числом -1, сверху данная последовательность ограниченна числом 0 (как  $\lim_{n \rightarrow +\infty}1/(1-n) = 0 $. Отсюда следует, что в целом $b_n$ ограничена.

\begin{equation}
b_{5} = \frac{1}{1-5} = - \frac{1}{4}
\end{equation}


**Заданине 3.1c**
\begin{equation}
\begin{gathered}
\left\{c_n\right\}_{n=1}^{\infty} = (-1)^n + \sqrt{2n}
\end{gathered}
\end{equation}

Данная последовательность не является монотонной. Этот факт несложно увидеть, рассмотрев отдельно четные и нечетные слагаемые:
\begin{equation}
\begin{gathered}
\left\{c_{2k}\right\}_{k=1}^{\infty} = 1 + 2\sqrt{k}, \\
\left\{c_{2k+1}\right\}_{k=0}^{\infty} = -1 + 2\sqrt{k+0.5}, \\
\end{gathered}
\end{equation}
При больших $k$ величиной 0.5 под корнем можно принебречь и мы видим, что четные слагаемые больше нечетных.

Данная последовательность ограничена снизу (при $n=1$) числом $ -1 + \sqrt{2}$, сверху данная последовательность неограниченна. Отсюда следует, что в целом $c_n$ неограничена.

\begin{equation}
c_{5} = -1 + \sqrt{10}
\end{equation}

Таким образом последовательность монотонно убывает.

Данная последовательность ограничена сверху (при $n=1$) числом $2$, снизу числом 1. Отсюда следует, что в целом $d_n$ ограничена.

### Задание 3.3 и 3.4

Предел последовательности
\begin{equation}
\lim_{n  \rightarrow +\infty} \frac{n}{\sqrt[n]{n!}}
\end{equation}
может быть записан в виде: 
\begin{equation}
\lim_{n  \rightarrow +\infty} \frac{n}{\sqrt[n]{n!}} = \lim_{n  \rightarrow +\infty} \sqrt[n]{\frac{n^n}{n!}} = \lim_{n  \rightarrow +\infty} \sqrt[n]{a_n},
\end{equation}
где $a_n = \frac{n^n}{n!}$. Поскольку $a_n > 0$ можем вопользоваться следующей теоремой: 
\begin{equation}
\lim_{n  \rightarrow +\infty} \sqrt[n]{a_n} = \lim_{n  \rightarrow +\infty} \frac{a_{n+1}}{a_n}
\end{equation}
В результате получим: 
\begin{equation}
\lim_{n  \rightarrow +\infty} \frac{n}{\sqrt[n]{n!}} = \lim_{n  \rightarrow +\infty} \sqrt[n]{\frac{n^n}{n!}} = \lim_{n  \rightarrow +\infty} \frac{(n+1)^{n+1}}{(n+1)!} \frac{n!}{n^n} = \lim_{n  \rightarrow +\infty} \frac{(n+1)^n}{n^n} = e,
\end{equation}

В силу того, что на Python вычисление факториала при больших значениях $n$ затруднительно (даже при использовании Decimal()), для достижения необходимой точность, воспользуемся предпоследней формулой, а именно:
\begin{equation}
\lim_{n  \rightarrow +\infty} \frac{n}{\sqrt[n]{n!}} = \lim_{n  \rightarrow +\infty} \frac{(n+1)^n}{n^n}.
\end{equation}

```python
%%time

import matplotlib
import math
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update({'font.size': 12})

import math

def series_n(n):
    return (1 + 1/n)**n

epsilon = 1e-7
n = 1
while abs(series_n(n + 1) - series_n(n)) > epsilon:
    n = n + 1
    
print('Величина n:', n, '\nЗначение a_n:', series_n(n))
```

Оптимизацию возможно построить за счет того, что мы каждый раз удваиваем величину $n$, далее как только мы достигнем нужной точности или ее превзойдем, можем в цикле в обратном порядке дойти до того $n$, с которого будет выполняться требуемое условие. Пример реализации ниже.

```python
%%time

import matplotlib
import math
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update({'font.size': 12})


def series_n(n):
    return math.pow((1 + 1/n), n)

epsilon = 1e-7
n = 1
root_x = [series_n(1)] #для отрисовки
root_y = [0]           #для отрисовки
while abs(series_n(n + 1) - series_n(n)) > epsilon:
    n = n * 2
    root_x.append(series_n(n)) #для отрисовки
    root_y.append(0)           #для отрисовки
    
while abs(series_n(n + 1) - series_n(n)) < epsilon:
    n = n - 1

print('Величина n:', n + 1, '\nЗначение a_n:', series_n(n))
```
